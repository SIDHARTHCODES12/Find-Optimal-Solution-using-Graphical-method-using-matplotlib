```python
#RegNo 2367413    Name:Nishant    Prog:Optimal solution using matplotlib(lab14)
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-2, 4, 400)
y = np.linspace(-2, 4, 400)
X, Y = np.meshgrid(x, y)
Z_obj = (X - 1)**2 + Y - 2
Z_con1 = Y - X - 1
Z_con2 = np.maximum(0, X + Y - 2)
plt.figure(figsize=(8, 6))
contour_obj = plt.contour(X, Y, Z_obj, levels=np.linspace(-.25, 15, 1), cmap='viridis')
contour_obj1 = plt.contour(X, Y, Z_obj, levels=np.linspace(-2, 10, 5), cmap='viridis')
plt.contour(X, Y, Z_con1, levels=[0], colors='r')
plt.contour(X, Y, Z_con2, levels=[0], colors='g')
plt.fill_between(x, -2, 2 - x, where=(2 - x) >= -2, interpolate=True, alpha=0.3, color='blue',
                 label='Feasible Region')
plt.scatter(0.5, 1.5, color='red', marker='o', s=50, label='Optimal Point',zorder=10)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Optimization with Constraints')
plt.legend()
plt.clabel(contour_obj, inline=True, fontsize=10)
plt.clabel(contour_obj1, inline=True, fontsize=8)
plt.grid(True)
plt.show()
```
